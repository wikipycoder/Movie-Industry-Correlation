```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use("ggplot")
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (12, 8)
import numpy as np
```

```python
df = pd.read_csv("movies.csv")
```

```python
df.head()
```

```python
for col in df.columns:
    missing_percent = np.mean(df[col].isnull())
    print("{} - {}%".format(col, missing_percent))
```

```python
df = df.dropna(subset=["gross"])
df["gross"] = df["gross"].astype("int64")

```

```python
df.head()
```

```python
df["yearcorrect"] = df["released"].astype(str).str[:4]
```

```python
df.sort_values(by=["gross"], inplace=False, ascending=False)
```

```python
df.head(100)
```

```python
pd.set_option("display.max_rows", None)
```

```python
plt.scatter(x=df["budget"], y=df["gross"])
plt.xlabel('Budget of the movie')
plt.ylabel('Gross of the movie')
plt.title('Budget and Gross Earnings')
plt.show()
```

```python
sns.regplot(x="budget", y="gross", data=df, scatter_kws={"color": "red"}, line_kws={"color": "green"})
plt.show()
```

```python
correlation_matrix = df[["budget", "gross", "runtime", "score", "votes", "year"]].corr(method="spearman")
```

```python
numerized_dataframe = df
for col in numerized_dataframe.columns:
    if numerized_dataframe[col].dtype == "category":
        numerized_dataframe[col] = numerized_dataframe[col].astype("category")
        numerized_dataframe[col] = numerized_dataframe[col].cat.codes
```

```python
correlation_matrix1 = numerized_dataframe.corr(method="spearman")
```

```python
sns.heatmap(correlation_matrix1, annot=True)
```
